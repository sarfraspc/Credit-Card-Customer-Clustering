## Correlation Heatmap  
Check multicollinearity and relationships that could affect clustering behavior.

```python
import seaborn as sns

plt.figure(figsize=(14,10))
sns.heatmap(df.corr(),annot=True, cmap='coolwarm')
plt.title('correlation map')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scale=StandardScaler()
df_scaled=scale.fit_transform(df)
```

### Reduce dimensionality for better visualization and efficient clustering without much info loss.

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
components = pca.fit_transform(df_scaled)
```

## Silhouette Scores for K-Means  
Evaluate different K values to determine optimal cluster count 

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

silhouette=[]
K = range(2, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(components)
    score = silhouette_score(components, labels)
    silhouette.append(score)
    print(f"{k}=Silhouette Score: {score:.3f}")
```

### After analyzing silhouette scores and observing cluster separation visually, **4 clusters** were selected for K-Means,mainly because getting clear actionable insights. eventhough its not got better score than 2 or 3

```python
km=KMeans(n_clusters=4,random_state=5)
df['km_cluster']=km.fit_predict(components)
```

```python
plt.plot(K,silhouette,marker='o')
plt.xlabel("Number of Clusters K")
plt.ylabel("Inertia")
plt.title("Silhouette Scores")
plt.grid()
plt.show()
```

```python

df['PCA1'], df['PCA2'] = components[:,0], components[:,1]

plt.figure(figsize=(10, 6))
plt.scatter(df['PCA1'], df['PCA2'], c=df['km_cluster'], cmap='tab10')
plt.title("Customer Segments - PCA-Reduced")
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.colorbar(label="Cluster")
plt.show()
```

### Let’s try clustering with DBSCAN to explore density-based patterns.

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
import numpy as np

neighbors = NearestNeighbors(n_neighbors=4)
neighbors_fit = neighbors.fit(components)
distances, indices = neighbors_fit.kneighbors(components)
distances = np.sort(distances[:, 3])

plt.plot(distances)
plt.title('Knee plot for DBSCAN eps')
plt.show()
```

```python
db=DBSCAN(eps=0.5, min_samples=4)
df['Cluster_DBSCAN'] = db.fit_predict(components)

print(silhouette_score(components,df['Cluster_DBSCAN']))
```

```python
df['Cluster_DBSCAN'].value_counts()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['PCA1'], df['PCA2'], c=df['Cluster_DBSCAN'], cmap='Set1', s=30)
plt.title("DBSCAN Clusters")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.colorbar(label='Cluster Label')
plt.show()
```
